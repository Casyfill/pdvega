# ``pdvega``: Pandas Plotting API to Vega-Lite

[``pdvega``](http://github.com/jakevdp/pdvega) is an experimental package that enables the creation of Vega-Lite plots using the pandas plotting API.

This is currently a proof-of-concept, and far from complete.

```python
import pandas as pd
import numpy as np
```

```python
# import adds the vgplot attribute to Series and DataFrames
import pdvega

hasattr(pd.DataFrame, 'vgplot')
```

```python
# Setup matplotlib for comparison
%matplotlib inline
import matplotlib.pyplot as plt
plt.style.use('seaborn-whitegrid')
```

## Line Plots

Here is an example of some time-series plotted as lines.
We'll use the [vega_datasets](https://github.com/jakevdp/vega_datasets) package for example data (install with ``pip install vega_datasets``)

```python
from vega_datasets import data

df = data.stocks(pivoted=True)
df.head()
```

```python
df.plot.line();
```

```python
df.vgplot.line()
```

## Scatter Plots

```python
df = data('cars')
df.head()
```

```python
df.plot.scatter(x='Horsepower', y='Miles_per_Gallon', c='Cylinders');
```

```python
df.vgplot.scatter(x='Horsepower', y='Miles_per_Gallon', c='Cylinders')
```

## Bar Plots

```python
rand = np.random.RandomState(42)
df = pd.DataFrame({'x': 0.5 + 0.5 * rand.rand(20),
                   'y': 0.7 + 0.5 * rand.rand(20)})
```

### Unstacked

```python
df.iloc[-10:].plot.bar();
```

```python
df.vgplot.bar()
```

### Stacked

```python
df.plot.bar(stacked=True);
```

```python
df.vgplot.bar(stacked=True)
```

### Horizontal bar

```python
df10 = df.iloc[:10]

df10.plot.barh()
```

```python
df10.vgplot.barh()
```

```python
df10.plot.barh(stacked=True);
```

```python
df10.vgplot.barh(stacked=True)
```

```python
df10.reset_index().plot.barh(x='index', y='y');
```

```python
df10.reset_index().vgplot.barh(x='index', y='y')
```

## Area Plots

```python
df = pd.DataFrame({'x': 1 + np.random.rand(50),
                   'y': 1 + np.random.rand(50)})
```

```python
df.plot.area();
```

```python
df.vgplot.area()
```

```python
df.plot.area(stacked=False);
```

```python
df.vgplot.area(stacked=False)
```
