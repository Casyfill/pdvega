# ``pdvega``: Pandas Plotting API to Vega-Lite

[``pdvega``](http://github.com/jakevdp/pdvega) is an experimental package that enables the creation of Vega-Lite plots using the pandas plotting API.

This is currently a proof-of-concept, and far from complete.

```python
# import adds the vgplot attribute to Series and DataFrames
import pdvega
```

## Line Plots

Here is an example of some time-series plotted as lines

```python
from vega_datasets import data
import pandas as pd
import numpy as np

df = data('stocks')
df['date'] = pd.to_datetime(df['date'])
df = df.pivot(index='date', columns='symbol', values='price')
df.head()
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
plt.style.use('seaborn-whitegrid')
```

```python
df.plot.line();
```

```python
df.vgplot.line()
```

## Scatter Plots

```python
df = data('cars')
df.head()
```

```python
df.plot.scatter(x='Horsepower', y='Miles_per_Gallon', c='Cylinders');
```

```python
df.vgplot.scatter(x='Horsepower', y='Miles_per_Gallon', c='Cylinders')
```

## Bar Plots

```python
df = pd.DataFrame({'x': 1 + 2 * np.random.rand(20),
                   'y': 0.5 + np.random.rand(20)})
```

### Unstacked

```python
df.iloc[-10:].plot.bar()
```

```python
df.vgplot.bar()
```

### Stacked

```python
df.plot.bar(stacked=True);
```

```python
df.vgplot.bar(stacked=True)
```

## Area Plots

```python
df = pd.DataFrame({'x': 1 + np.random.rand(50),
                   'y': 1 + np.random.rand(50)})
```

```python
df.plot.area();
```

```python
df.vgplot.area()
```

```python
df.plot.area(stacked=False);
```

```python
df.vgplot.area(stacked=False)
```
