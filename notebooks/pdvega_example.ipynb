# ``pdvega``: Pandas Plotting API to Vega-Lite

``pdvega`` is an experimental package that enables the creation of Vega-Lite plots using the pandas plotting API.

See http://github.com/jakevdp/pdvega

This is currently a proof-of-concept, and far from complete.

```python
# import adds the vgplot attribute to Series and DataFrames
import pdvega

import pandas as pd
import numpy as np

t = np.arange(100)
df = pd.DataFrame({'x': np.sin(t / 10),
                   'y': np.cos(t / 10)},
                  index=t)
```

## Line Plots

```python
%matplotlib inline
df.plot.line();
```

```python
df.vgplot.line()
```

## Scatter Plots

```python
df.plot.scatter(x='x', y='y');
```

```python
df.vgplot.scatter(x='x', y='y')
```

## Bar Plots

```python
df.iloc[5:15].plot.bar();
```

```python
df.iloc[5:15].vgplot.bar()
```
